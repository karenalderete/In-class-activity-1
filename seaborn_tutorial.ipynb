```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seabon as sns
```

Load a sample dataset 

Seaborn comes with built-in datasets. We’ll use the `tips` dataset, which contains restaurant bills and tips (great for exploring relationships between numerical and categorical variables).


```python
tips = sns.load_dataset("tips")
tips.head()
```

Create a Histogram with KDE (Basic Plot)

We’ll plot the distribution of `total_bill` to see how the data is spread out. We’ll also add a Kernel Density Estimate (KDE) curve to better understand the shape of the distribution.

```python
sns.histplot(data=tips, x="total_bill", kde=True)
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Count")
plt.show()
```

Create a Box Plot (Categorical Plot)

Boxplots are great for comparing distributions across categorical groups.
Here, we’ll look at how `total_bill` varies by day of the week and by gender.

The following box plot will help us analyze keypoints such as how median and IQR of bills change by day; whether men or women tip more (on average); if there are any outliers (values that are significantly higher or lower).


```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Total Bill Distribution by Day and Gender")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.legend(title="Gender")
plt.show()
```

Create a Relational Plot (Scatterplot with Regression Line)
 
Scatterplots help visualize the relationship between two numeric variables. 
Using Seaborn’s `sns.lmplot()`, we can also add a regression line to see general trends (perfect for linear relationships).

The following visualization will show us keypoints such as how tips increase with total bill (positive correlation); whether men or women tend to tip more per dollar; if the relationship is linear or not.

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex", aspect=1.5)
plt.title("Relationship Between Total Bill and Tip by Gender")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

Create an Advanced Plot (Heatmap)

Heatmaps are great for visualizing relationships between two categorical variables (like time and place) 
with a third variable represented by color intensity. In this example, we’ll visualize how airline passenger volume
changes over time using the `flights` dataset.


```python
# Load and pivot the dataset
flights = sns.load_dataset("flights").pivot(index="month", columns="year", values="passengers")

# Plot heatmap
sns.heatmap(flights, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Monthly Airline Passengers (1949–1960)")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

Create a Pairplot - Exploring Relationships Across Multiple Variables BONUS :)


A `pairplot` creates a grid of scatterplots and histograms to show pairwise relationship
s between all numeric variables in a dataset. It’s ideal for identifying trends, correlations, and clusters at a glance.

HeI, we’ll use the `tips` dataset again to explore patterns based on gender.

```python
sns.pairplot(data=tips, hue="sex")
plt.suptitle("Pairplot of Numeric Variables in Tips Dataset", y=1.02)
plt.show()
```

Styling with Themes

Seaborn allows us to change the style of our plots with just one line of code. This is very useful for making charts
presentation-ready or matching a specific visual aesthetic.

There are options like:
"darkgrid" (default)

"whitegrid" (great for presentations)

"white" (clean look)

"dark" (good for dark backgrounds)

"ticks" (subtle axis ticks)

```python
# Set the visual theme
sns.set_style("whitegrid")  # Options: darkgrid, whitegrid, dark, white, ticks

# Example plot with the new theme
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", palette="pastel")
plt.title("Styled Boxplot with Whitegrid Theme")
plt.show()

```
